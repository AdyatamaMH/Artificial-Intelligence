```python
import pandas as pd
from transformers import AutoTokenizer


df = pd.read_csv('Data_latih.csv')
display(df.head())
```

```python
df = df.rename(columns={
    'ID': 'record_id',
    'judul': 'title',
    'narasi': 'content',
    'tanggal': 'publication_date'
})

display(df.head())
```

```python
df = df.drop(columns=['nama file gambar', 'label'])
display(df.head())
```

```python
df = df[['record_id', 'title', 'content', 'publication_date']]
display(df.head())
```

```python
text_columns = ['record_id', 'title', 'content', 'publication_date']
for col in text_columns:
    df[col] = df[col].apply(lambda x: x.lower() if isinstance(x, str) else x)
display(df.head())
```

```python
df['Status'] = 0

print(df)
```

```python
df = df.iloc[1:].reset_index(drop=True)
df.to_csv('cleaned_data_hoax.csv', index=False)
display(df.tail())
```

```python
cleaned_file_path = 'cleaned_data_hoax.csv'
df = pd.read_csv(cleaned_file_path)

columns_to_tokenize = ['title', 'content']

tokenizer = AutoTokenizer.from_pretrained("allenai/longformer-base-4096")

for column in columns_to_tokenize:
    df[f"{column}_tokens"] = df[column].apply(
        lambda x: tokenizer.tokenize(x) if pd.notnull(x) else []
    )
    df[f"{column}_token_ids"] = df[column].apply(
        lambda x: tokenizer.encode(x, truncation=True, max_length=512) if pd.notnull(x) else []
    )

tokenized_file_path = 'tokenized_data_hoax.csv'
df.to_csv(tokenized_file_path, index=False)
print(f"Tokenized data saved to {tokenized_file_path}")
```

```python
def reconstruction_accuracy(original_texts, tokenized_texts, tokenizer):
    matches = 0
    total = len(original_texts)
    for original, tokens in zip(original_texts, tokenized_texts):
        reconstructed = tokenizer.convert_tokens_to_string(tokens)
        if original == reconstructed:
            matches += 1
    return matches / total
for column in columns_to_tokenize:
    sample_text = df[column].iloc[0]
    tokens = tokenizer.tokenize(sample_text)
    print(f"Original: {sample_text}")
    print(f"Tokens: {tokens}")

```
