```python
import pandas as pd

df = pd.read_csv('Data_latih.csv')
display(df.head())
```

```python
df = df.rename(columns={
    'ID': 'record_id',
    'judul': 'title',
    'narasi': 'content',
    'tanggal': 'publication_date'
})

display(df.head())
```

```python
df = df.drop(columns=['nama file gambar', 'label'])
display(df.head())
```

```python
df = df[['record_id', 'title', 'content', 'publication_date']]
display(df.head())
```

```python
text_columns = ['record_id', 'title', 'content', 'publication_date']
for col in text_columns:
    df[col] = df[col].apply(lambda x: x.lower() if isinstance(x, str) else x)
display(df.head())
```

```python
df = df.iloc[1:].reset_index(drop=True)
df.to_csv('cleaned_data_hoax.csv', index=False)
display(df.head())
```
